```python
import numpy as np 
import matplotlib.pyplot as plt 

n = 400
sigma1 = 1
mu1 = 0 
sigma2 = 4
mu2 = 0 

x1 = np.random.uniform(-np.sqrt(3),np.sqrt(3),n)
x2 = np.random.normal(mu2,sigma2,n) 

plt.scatter(x1,x2)
plt.show()
```

```python
x2_exp = np.random.exponential(1,n)
x2_e_ctr = x2_exp - 1

plt.scatter(x1,x2_e_ctr)
plt.show()
```

```python
Q = lambda theta: np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
theta= np.pi/4
Y = Q(theta) @ np.array([x1,x2_e_ctr])
plt.scatter(Y[0],Y[1])
```
